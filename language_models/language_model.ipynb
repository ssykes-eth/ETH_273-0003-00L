# Basic Language Model in Python

This notebook demonstrates how to create a basic language model using Python. The model processes a small list of strings (vocabulary) and computes a language model, which is essentially a table showing the probability that a word follows a prefix in the input.

## 1. Import Required Libraries

```python
import random
```

## 2. Define Vocabulary and Input Data

```python
vocabulary = ['I', 'am', 'learning', 'Python', 'and', 'AI']
input_data = ['I am learning Python', 'I am learning AI']
```

## 3. Tokenize Input Data

```python
input_data_tokenized = [sentence.split() for sentence in input_data]
```

```python
input_data_tokenized
```

## 4. Create Prefix-Word Table

```python
prefix_word_table = {}
for sentence in input_data_tokenized:
    for i in range(len(sentence) - 1):
        prefix = sentence[i]
        word = sentence[i + 1]
        if prefix not in prefix_word_table:
            prefix_word_table[prefix] = {}
        if word not in prefix_word_table[prefix]:
            prefix_word_table[prefix][word] = 0
        prefix_word_table[prefix][word] += 1
```

```python
prefix_word_table
```

## 5. Normalize Prefix-Word Table

```python
for prefix, word_counts in prefix_word_table.items():
    total_count = sum(word_counts.values())
    for word, count in word_counts.items():
        prefix_word_table[prefix][word] = count / total_count
```

## 6. Generate Sentence from Prefix

```python
def generate_sentence(prefix):
    sentence = [prefix]
    while True:
        last_word = sentence[-1]
        if last_word not in prefix_word_table:
            break
        next_word_probs = prefix_word_table[last_word]
        words = list(next_word_probs.keys())
        probs = list(next_word_probs.values())
        cumulative = 0
        r = random.random()
        for word, p in zip(words, probs):
            cumulative += p
            if r <= cumulative:
                next_word = word
                break
        sentence.append(next_word)
    return ' '.join(sentence)

# Example usage:
print(generate_sentence('I'))
```
