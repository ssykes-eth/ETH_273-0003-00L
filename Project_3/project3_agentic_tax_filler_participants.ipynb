## 🏗️ MCP: How the agent uses the browser and the ZHPrivateTax simulator

In this project, your agent does not interact with the tax portal through hardcoded scripts or fixed coordinates. Instead, it utilizes the **Model Context Protocol (MCP)**. MCP is a standardized and open-source protocol, developed by Anthropic, that acts as an "universal translator" that allows an AI model to safely and effectively use tools provided by an external environment.

### The Technical Challenge: UI Auto-Discovery
One of the most powerful aspects of this project is **UI Auto-Discovery**. Unlike traditional web scrapers that rely on specific, fragile backend IDs (which break the moment a developer changes a line of code), your agent must "see" the page more like a human does in order to be able to fill it automatically.

In order to do this, the ZHPrivateTax Light simulator that we have built exposes an MCP server: a collection of tools like `fill_field()` that your AI agent can call directly to perform a certain action in a predictable, controllable way. The agent acts as an MCP client and talks to the simulator via the universal interface of Model Context Protocol. This is the same technology that state-of-the-art tools like Claude Code or ChatGPT desktop are using! This enables:

* **Semantic Scanning**: Your agent doesn't know where the "Gross Salary" field is until it arrives on the page. It must use MCP to scan the UI and identify fields based on their visible labels and roles.
* **Dynamic Interaction**: As the agent navigates the 6-page **ZHprivateTax "Light"** simulator, it receives a real-time map of the current page. It must then decide, on the fly, which field matches the data it extracted from the taxpayer's documents.

### Your Tools (The MCP Interface)
One of the good things about MCP is that you don't need to write the low-level code that controls the browser; that is handled by the **MCP Server** running behind the scenes in this Colab notebook and that exposes the tools. Instead, your task is to orchestrate the agent's logic to call these tools at the right time:

* **`scan_page()`**: This is the agent's "eyes." It returns a JSON representation of every input, button, and dropdown visible on the current screen, along with a "locator" handle for each.
* **`fill_field(locator, value)`**: This is the agent's "hands." Once a field is identified (e.g., the locator for "Date of Birth"), the agent uses this tool to input the correct information.
* **`click_element(locator)`**: This allows the agent to interact with the portal—clicking "Next Page," adding rows for new children, or selecting a "Yes/No" toggle for the subsidized canteen.
* **`submit_form()`**: The final action that packages the entire tax return into a JSON format for submission and scoring.

## 🧠 The Agent Loop — `think()`
> This is the brain of the agent.
> 1. Pre-reads all documents and guides (zero LLM calls)
> 2. Navigates through each form page (code-driven, no LLM needed)
> 3. For each page: sends fields + documents + guides to the LLM
> 4. The LLM returns `{"field-id": "value"}` — the code fills the fields
> 5. Submits when the review page is reached
>

````python
# ═══════════════════════════════════════════════════════════════════════
# 🧠 THE AGENT LOOP — think()
# ═══════════════════════════════════════════════════════════════════════
#
# This is the brain of the agent. Here's what it does:
#
#   1. Pre-reads all documents (zero LLM calls — just file I/O)
#   2. Pre-loads all tax guides (zero LLM calls — just file I/O)
#   3. Navigates through each form page (code-driven, no LLM needed)
#   4. For each page: sends fields + docs + guides to the LLM
#   5. LLM returns {field_id: value} → code fills the fields
#   6. Clicks "Next" → repeats until Review page → submits
#
# KEY DESIGN PRINCIPLE:
#   The code handles ALL navigation. The LLM's ONLY job is:
#   "Given this page's fields and all the documents, which values
#   go in which fields?"
#
# ─── AVAILABLE TOOLS (server methods) ──────────────────────
#
#   BROWSER TOOLS — interact with the tax portal:
#     server.scan_page()                → dict with page_name, elements[]
#     server.fill_field(locator, value) → dict with success, error
#     server.click_element(locator)     → dict with success, new_page
#     server.submit_form()              → dict with success, submission_json
#
#   KNOWLEDGE TOOLS — read files and guides:
#     server.list_documents()           → list of filenames (strings)
#     server.read_document(filepath)    → dict with content, type, structured
#     server.list_guides()              → list of guide dicts (title, url)
#     server.fetch_guide(url)           → dict with content, title, success
#
#   HUMAN-IN-THE-LOOP:
#     server.ask_user(question)         → dict with answer (⚠ penalised!)
#
# ─── WHAT YOU IMPLEMENT ────────────────────────────────────
#   🎯 TODO 6:  Pre-read all persona documents
#   🎯 TODO 7:  Auto-navigate to the first form page
#   🎯 TODO 8:  Build the LLM message (ask_llm_for_page)
#   🎯 TODO 9:  Fill the form fields from the LLM response
#   🎯 TODO 10: (Optional) Add verification / error recovery
#
# ─── WHAT'S ALREADY DONE FOR YOU ───────────────────────────
#   - Guide pre-loading
#   - State tracking
#   - Dynamic row handling (add_rows_if_needed)
#   - Main navigation loop (scan → fill → click next)
#   - Logging
#
# ═══════════════════════════════════════════════════════════════════════

import json
import time
import re


def think(server, client, system_prompt_ignored=None, max_steps=100):
    """Code-driven navigation + per-page LLM field mapping.

    Parameters
    ----------
    server : MCPServer
        The tool server — call server.scan_page(), server.fill_field(), etc.
    client : OpenAI
        The LLM client — call client.chat.completions.create(...)
    max_steps : int
        Safety limit to prevent infinite loops.

    Returns
    -------
    dict with keys: "submission", "filled_history", "log"
    """

    # ── Logging helper (prints AND records) ───────────────
    log_lines = []
    def log(msg):
        print(msg)
        log_lines.append(msg)


    # ══════════════════════════════════════════════════════
    # 🎯 TODO 6: PRE-READ ALL DOCUMENTS
    # ══════════════════════════════════════════════════════
    #
    # Before visiting any page, read ALL the persona's documents
    # so we can send them to the LLM on every page.
    #
    # WHY? The LLM doesn't decide what to read. We give it
    # everything upfront — the Lohnausweis, bank statement,
    # Pillar 3a confirmation, profile, receipts — all at once.
    #
    # STEPS:
    #   1. Call server.list_documents() to get all filenames
    #      Returns: ["bank_statement.csv", "lohnausweis.txt", ...]
    #
    #   2. For each filename, call server.read_document(filepath)
    #      Returns: {
    #          "filepath": "lohnausweis.txt",
    #          "type": "text",          # or "csv", "pdf", "json"
    #          "content": "=== LOHNAUSWEIS ===\nBruttolohn: 142,000...",
    #          "structured": null       # or parsed data for CSV/JSON
    #      }
    #
    #   3. Store each document in a dict: all_docs[filepath] = doc
    #
    #   4. Format ALL documents into a single text string (docs_text)
    #      that we'll inject into every LLM call. Example format:
    #
    #      === lohnausweis.txt ===
    #      Bruttolohn: 142,000 ...
    #
    #      === bank_statement.csv ===
    #      Date,Description,Amount,Balance
    #      2025-01-31,Lohn Google,+11833.33,45230.50
    #      ...
    #
    # EXAMPLE IMPLEMENTATION:
    #
    #   log("📂 Pre-reading all documents...")
    #   doc_list = server.list_documents()
    #   all_docs = {}
    #   for filepath in doc_list:
    #       doc = server.read_document(filepath)
    #       all_docs[filepath] = doc
    #       log(f"  📄 {filepath} ({doc['type']}, {len(doc.get('content', ''))} chars)")
    #
    #   docs_text = "\n\n".join(
    #       f"=== {name} ===\n{doc['content'][:8000]}"
    #       for name, doc in all_docs.items()
    #   )
    #   log(f"✅ Pre-read {len(all_docs)} documents.\n")
    #
    # ─── Your code below ──────────────────────────────────

    all_docs = {}
    docs_text = ""  # 🎯 IMPLEMENT: read all documents and build docs_text

    # ──────────────────────────────────────────────────────


    # ══════════════════════════════════════════════════════
    # PHASE 2: PRE-LOAD TAX GUIDES (complete — no changes)
    # ══════════════════════════════════════════════════════
    # The guides contain deduction rules, thresholds, and
    # formulas. They are loaded once and injected into the
    # LLM context for the relevant page.

    log("📚 Pre-loading tax guides...")
    all_guides = load_all_guides(
        "/content/project3-agentic-tax-filler/guides"
    )
    log(f"✅ Loaded {len(all_guides)} guides: {list(all_guides.keys())}")


    # ══════════════════════════════════════════════════════
    # PHASE 3: STATE TRACKING (complete — no changes)
    # ══════════════════════════════════════════════════════

    filled_history = {}   # {page_name: {locator: value, ...}}
    llm_calls = 0
    total_fields_filled = 0


    # ══════════════════════════════════════════════════════
    # 🎯 TODO 7: AUTO-NAVIGATE TO FIRST FORM PAGE
    # ══════════════════════════════════════════════════════
    #
    # The app starts on a root/overview page with no form fields.
    # We need to click into the actual form before the main loop.
    #
    # STEPS:
    #   1. Call server.scan_page() to see where we are
    #      Returns: {
    #          "page_name": "root",     ← we're on the overview
    #          "elements": [...],       ← buttons, not form fields
    #          "validation_errors": []
    #      }
    #
    #   2. Check if we're on a non-form page:
    #      if page_name is "root", "overview", or ""
    #
    #   3. Click navigation buttons to enter the form.
    #      The portal has buttons with these IDs:
    #        "tab-nav-form"      → switches to the form view
    #        "btn-nav-personal"  → goes to the personal page
    #        "btn-login"         → passes the login screen
    #
    #      Use server.click_element(locator) to click them.
    #      Returns: {
    #          "success": true,
    #          "new_page": "personal"   ← where we ended up
    #      }
    #
    #      Try each button in order. Stop when one succeeds
    #      and navigates to a new page.
    #
    # EXAMPLE IMPLEMENTATION:
    #
    #   initial = server.scan_page()
    #   page_name = initial.get("page_name", "")
    #   if not initial.get("elements") or page_name in ("root", "overview", ""):
    #       log("⚡ Auto-navigating to form...")
    #       for locator in ["tab-nav-form", "btn-nav-personal", "btn-login"]:
    #           result = server.click_element(locator)
    #           if result.get("success") and result.get("new_page"):
    #               log(f"  Navigated to: {result['new_page']}")
    #               break
    #
    # ─── Your code below ──────────────────────────────────

    pass  # 🎯 IMPLEMENT: scan the page, check if on overview, click into the form

    # ──────────────────────────────────────────────────────

    def _get_npc_hints(section: str) -> str:
        """Return targeted ask_user hints for fields known to fail in this section."""
        notes_path = Path(REPO) / "personas" / PERSONA / "private_notes.json"
        if not notes_path.exists():
            return ""
        try:
            with open(notes_path) as f:
                private_notes = json.load(f)
        except Exception:
            return ""
        clarifications = private_notes.get("clarifications", [])
        relevant = []
        for c in clarifications:
            field = c.get("field", "")
            if not field.startswith(section + "."):
                continue
            patterns = c.get("question_patterns", [])
            question = patterns[0] if patterns else field
            answer_hint = c.get("answer", "")
            relevant.append(
                f"- Field `{field}`: ALWAYS ask the taxpayer to confirm this field — "
                f"documents alone may not have all the information needed. "
                f"Suggested question: '{question}'"
            )
        if not relevant:
            return ""
        return "\n## Fields to verify with taxpayer (ALWAYS use ask_user for these fields)\n" + "\n".join(relevant) + "\n"

    # ══════════════════════════════════════════════════════
    # HELPER: ask_llm_for_page
    # ══════════════════════════════════════════════════════
    #
    # This function builds the full message for the LLM and
    # sends it. The LLM's response is parsed into a dict of
    # {field_locator: value} pairs.
    #
    # The message has 5 components:
    #   1. System prompt = BASE_PROMPT + SECTION_PROMPTS[section]
    #   2. All taxpayer documents (docs_text)
    #   3. History of previously filled fields
    #   4. Relevant tax guides for this page
    #   5. Current page's field descriptions
    #
    def ask_llm_for_page(page_name, fields):
        nonlocal llm_calls

        # ─── Step A: Build the system message ─────────
        # Combine BASE_PROMPT with the section-specific prompt.
        #
        # The section is the first part of the page name:
        #   "deductions/professional" → "deductions"
        #   "personal/children" → "personal"
        #   "income" → "income"

        section = page_name.split("/")[0]
        section_prompt = SECTION_PROMPTS.get(section, "")
        system_msg = BASE_PROMPT + "\n" + section_prompt


        # ─── Step B: Build the history summary ────────
        # Tell the LLM what we already filled on previous pages.
        # This prevents duplicate entries and helps with context
        # (e.g., if income was 142,000 on the income page, the
        # deductions page can reference that for calculations).

        history_lines = []
        for prev_page, prev_fields in filled_history.items():
            entries = ", ".join(
                f"{k}={v}" for k, v in prev_fields.items()
                if not k.startswith("_")
            )
            if entries:
                history_lines.append(f"  {prev_page}: {entries}")
        history_text = "\n".join(history_lines) if history_lines else "  (none yet)"


        # ─── Step C: Filter to fillable fields ────────
        # Remove buttons and display text — only keep actual
        # input fields (text inputs, selects, checkboxes, etc.)

        fillable = [f for f in fields if f.get("type") not in ("button", "text")]
        if not fillable:
            return {}

        fields_desc = json.dumps(fillable, indent=2)


        # ─── Step D: Get relevant tax guides ──────────
        # The get_guides_for_page function returns the guide
        # text relevant to the current page type.

        guide_text = get_guides_for_page(page_name, all_guides)
        guides_section = ""
        if guide_text or POLICY_KB_TEXT:
            guides_section = f"""

## TAX GUIDES (consult these for rules, thresholds, and formulas)
{POLICY_KB_TEXT}

{guide_text}
"""


        # ═══════════════════════════════════════════════
        # 🎯 TODO 8: Build the user message and call the LLM
        # ═══════════════════════════════════════════════
        #
        # You need to:
        #   1. Assemble the user message from the components above
        #   2. Call the LLM with system_msg and user_msg
        #   3. Parse the response into a {locator: value} dict
        #
        # COMPONENTS AVAILABLE TO YOU:
        #   - docs_text       → all documents as a single string
        #   - history_text    → what was filled on previous pages
        #   - guides_section  → relevant tax guides for this page
        #   - page_name       → e.g., "deductions/professional"
        #   - fields_desc     → JSON of fillable fields on this page
        #
        # EXAMPLE of what the user message should look like:
        #
        #   ## Taxpayer Documents
        #   === lohnausweis.txt ===
        #   Bruttolohn: 142,000 ...
        #
        #   ## Previously Filled
        #     personal: field-...-firstName=Anna, ...
        #
        #   ## TAX GUIDES
        #   (guide text here)
        #
        #   ## Current Page: deductions/professional
        #   Fields on this page:
        #   [{"type": "input", "locator": "field-deductions-...", ...}]
        #
        #   Return a JSON object mapping field locators to values.
        #
        # HOW TO CALL THE LLM:
        #
        #   response = client.chat.completions.create(
        #       model=LLM_MODEL,
        #       max_tokens=4096,
        #       messages=[
        #           {"role": "system", "content": system_msg},
        #           {"role": "user",   "content": user_msg}
        #       ]
        #   )
        #   raw_text = response.choices[0].message.content
        #
        # HOW TO PARSE THE RESPONSE:
        #   Use the provided helper:
        #       parse_field_mapping(raw_text)
        #   This extracts a JSON dict from the LLM's response text,
        #   handling markdown code fences and other formatting.
        #
        # RATE LIMITING:
        #   Add time.sleep(5) before the API call to avoid hitting
        #   rate limits. If you get a 429 error, wait longer (45s)
        #   and retry.
        #
        # ─── Your code below ──────────────────────────

        # Step 1: Assemble the user message
        user_msg = ""  # 🎯 BUILD THIS — combine docs_text, history_text,
                       #    guides_section, page_name, and fields_desc
                       #    into a single formatted string

        # Step 2: Rate limit protection
        time.sleep(5)

        # Step 3: Call the LLM
        # response = ...  # 🎯 CALL client.chat.completions.create(...)

        # Step 4: Parse the response
        llm_calls += 1
        # return parse_field_mapping(response.choices[0].message.content)  # 🎯 UNCOMMENT
        return {}  # ← Remove this once you implement the above


    # ══════════════════════════════════════════════════════
    # HELPER: fill_page
    # ══════════════════════════════════════════════════════

    def fill_page(mapping, page_name):
        """Fill form fields from the LLM's {locator: value} mapping.

        Parameters
        ----------
        mapping : dict
            Keys are field locators (e.g., "field-income-employment-bruttolohn")
            Values are the values to fill (e.g., "142000")
            Special keys starting with "_" are metadata (e.g., "_thought")

        page_name : str
            The current page name for logging

        HOW FILLING WORKS:
            Call server.fill_field(locator, value) for each entry.
            This calls the browser bridge which executes:
                window.TaxPortal.fillField("field-income-...", "142000")
            in the React app. The field gets filled, and the browser
            returns {success: true/false, error: ...}.
        """
        nonlocal total_fields_filled
        filled = {}

        # ═══════════════════════════════════════════════
        # 🎯 TODO 9: Fill each field from the mapping
        # ═══════════════════════════════════════════════
        #
        # You need to:
        #   1. Extract and log the "_thought" key (LLM's reasoning)
        #   2. Loop through the remaining keys in mapping
        #   3. Skip keys that start with "_" (metadata, not fields)
        #   4. For each field, call:
        #          server.fill_field(locator, str(value))
        #      This returns: {"success": True/False, "error": "..."}
        #   5. Check if result.get("success") is True
        #   6. Log success (✅) or failure (❌)
        #   7. Track what was filled in the `filled` dict
        #
        # EXAMPLE:
        #
        #   # Extract reasoning
        #   thought = mapping.pop("_thought", None)
        #   if thought:
        #       log(f"  🧠 {thought}")
        #
        #   # Fill each field
        #   for locator, value in mapping.items():
        #       if locator.startswith("_"):
        #           continue  # skip metadata keys like _rows_needed
        #       result = server.fill_field(locator, str(value))
        #       if result.get("success"):
        #           filled[locator] = value
        #           total_fields_filled += 1
        #           log(f"  ✅ {locator} = {value}")
        #       else:
        #           log(f"  ❌ {locator}: {result.get('error', 'unknown')}")
        #
        # ─── Your code below ──────────────────────────

        pass  # 🎯 REPLACE with your implementation

        # ──────────────────────────────────────────────

        # Record what was filled for this page (used by history)
        if filled:
            filled_history[page_name] = filled
        return filled


    # ══════════════════════════════════════════════════════
    # HELPER: add_rows_if_needed (COMPLETE — no changes needed)
    # ══════════════════════════════════════════════════════

    def add_rows_if_needed(page_name, elements, num_rows_needed=0):
        """Handle dynamic table rows (children, bank accounts, etc.).

        Some form sections start empty — no input fields until you
        click '+ Add Row'. This function detects add-row buttons,
        counts existing rows, and adds more if needed.

        HOW IT WORKS:
            The "Add Row" buttons have IDs like:
                "btn-add-row-personal-children"
                "btn-add-row-wealth-bankaccounts"
            Clicking them with server.click_element() creates new
            fields with indexed IDs:
                "field-personal-children-0-name"
                "field-personal-children-1-name"
            The -0-, -1- index tells us how many rows exist.
        """
        add_btns = [e for e in elements
                    if e.get("type") == "button"
                    and "add-row" in (e.get("locator") or "").lower()]

        if not add_btns:
            return None

        existing_rows = set()
        for e in elements:
            loc = e.get("locator") or ""
            match = re.search(r'-(\d+)-', loc)
            if match:
                existing_rows.add(int(match.group(1)))
        num_existing = len(existing_rows)

        rows_to_add = max(0, (num_rows_needed or 1) - num_existing)

        if rows_to_add > 0:
            for _ in range(rows_to_add):
                for btn in add_btns:
                    loc = btn.get("locator")
                    if loc:
                        log(f"  ➕ Adding row: {loc}")
                        server.click_element(loc)
            return server.scan_page()

        return None


    # ══════════════════════════════════════════════════════
    # MAIN NAVIGATION LOOP
    # ══════════════════════════════════════════════════════
    #
    # This loop visits every page in the form, calls the LLM
    # for each page, fills the fields, and moves to the next.
    #
    # HOW NAVIGATION WORKS:
    #   - Every form page has a "Next" button with ID "btn-next"
    #     → server.click_element("btn-next") moves to the next page
    #   - The sidebar has section buttons:
    #     → "nav-personal", "nav-income", "nav-deductions",
    #       "nav-wealth", "nav-attachments", "nav-review"
    #   - The top bar has "tab-nav-form" to switch to form view
    #   - The code clicks these buttons — the LLM NEVER decides
    #     where to navigate
    #
    # LOOP STRUCTURE:
    #   1. server.scan_page()        → see what's on this page
    #   2. Skip if overview or already visited
    #   3. server.submit_form()      → if on review page
    #   4. add_rows_if_needed()      → for dynamic tables
    #   5. ask_llm_for_page()        → LLM decides field values
    #   6. fill_page()               → fill the fields
    #   7. server.click_element("btn-next") → go to next page
    #

    log("🚀 Starting code-driven navigation...\n")
    step = 0
    visited = set()

    while step < max_steps:
        step += 1

        # ─── Scan: see what's on this page ────────────
        page_data = server.scan_page()
        page_name = page_data.get("page_name", "unknown")
        elements = page_data.get("elements", [])

        # Skip overview/root pages
        if page_name in ("root", "overview", ""):
            server.click_element("tab-nav-form")
            continue

        # Skip already-visited pages (prevent infinite loops)
        if page_name in visited:
            server.click_element("btn-next")  # "btn-next" = the Next button ID
            continue
        visited.add(page_name)

        log(f"--- Page {step}: {page_name} ---")

        # ─── Submit on the review page ────────────────
        if page_name == "review":
            log("📋 Review page — submitting...")
            result = server.submit_form()
            log(f"  ✅ Submitted!")
            log(f"\n📊 Summary: {llm_calls} LLM calls, {total_fields_filled} fields filled")
            return {
                "submission": result,
                "filled_history": filled_history,
                "log": log_lines,
            }

        # ─── Count fillable fields ────────────────────
        fillable = [e for e in elements if e.get("type") not in ("button", "text")]
        has_add_row = any(
            e.get("type") == "button" and "add-row" in (e.get("locator") or "").lower()
            for e in elements
        )

        if not fillable and not has_add_row:
            log(f"  ⏭️ No fillable fields — skipping")
        else:
            # If dynamic section with no rows yet, add 1 so LLM sees fields
            if not fillable and has_add_row:
                new_scan = add_rows_if_needed(page_name, elements, 1)
                if new_scan:
                    elements = new_scan.get("elements", [])
                    fillable = [e for e in elements if e.get("type") not in ("button", "text")]

            log(f"  📝 {len(fillable)} fillable fields")

            # ─── Ask LLM what to fill on this page ────
            mapping = ask_llm_for_page(page_name, elements)

            # Handle _ask_user requests from main LLM (NPC loop)
            ask_user_iterations = 0
            while "_ask_user" in mapping and ask_user_iterations < 3:
                ask_user_iterations += 1
                question = mapping.pop("_ask_user")
                if isinstance(question, list):
                    question = question[0]
                if not question:
                    log("  ⚠️ Empty question from LLM, skipping NPC call")
                    break
                log(f"  ❓ Main LLM asks: {question}")

                # Phase 1: show question + typing indicator in the popup
                try:
                    server._bridge.notify_ask_user(question, "")
                except Exception:
                    pass

                result = server.ask_user(question)
                answer = result["answer"]
                log(f"  💬 NPC answers: {answer}")

                # Phase 2: update popup with the actual answer (typewriter effect)
                try:
                    server._bridge.notify_ask_user(question, answer)
                    time.sleep(10)  # pause so user can watch the typing animation
                except Exception:
                    pass

                # Append the Q&A to the document context
                docs_text += f"\n\n=== Taxpayer Answer ===\nQ: {question}\nA: {answer}"

                # Re-ask main LLM with updated context (includes the answer)
                mapping = ask_llm_for_page(page_name, elements)

            if mapping:
                # Handle dynamic rows — LLM says "_rows_needed": N
                rows_needed = int(mapping.pop("_rows_needed", 0))
                if rows_needed > 1:
                    new_scan = add_rows_if_needed(page_name, elements, rows_needed)
                    if new_scan:
                        elements = new_scan.get("elements", [])
                        mapping = ask_llm_for_page(page_name, elements)
                        if mapping:
                            mapping.pop("_rows_needed", None)

                fill_page(mapping, page_name)
            else:
                log(f"  ⏭️ LLM returned empty — nothing to fill")

        # ─── Navigate to the next page ────────────────
        # Every form page has a "Next" button with ID "btn-next".
        # We click it to move forward. This is code-driven —
        # the LLM never decides which page to go to.
        nav_result = server.click_element("btn-next")
        if nav_result.get("success") and nav_result.get("new_page"):
            log(f"  ➡️ Navigated to: {nav_result['new_page']}")
        else:
            # Fallback: if "btn-next" doesn't work (e.g., last page
            # in a section), try the sidebar navigation buttons.
            # Sidebar button IDs: "nav-personal", "nav-income",
            # "nav-deductions", "nav-wealth", "nav-attachments",
            # "nav-review". Top bar: "tab-nav-form".
            for nav_target in ["tab-nav-form", "nav-personal", "nav-income",
                              "nav-deductions", "nav-wealth",
                              "nav-attachments", "nav-review"]:
                r = server.click_element(nav_target)
                if r.get("success") and r.get("new_page"):
                    log(f"  ➡️ Sidebar jump to: {r['new_page']}")
                    break
            else:
                log(f"  ❌ Cannot navigate further — stopping")
                break

    log(f"⚠️ Max steps ({max_steps}) reached without submitting.")
    return {"submission": None, "filled_history": filled_history, "log": log_lines}
````

````python
# ═══════════════════════════════════════════════════════════════════════
# 🧠 FULLY AGENTIC think() — The LLM decides every action
# ═══════════════════════════════════════════════════════════════════════
#
# This version lets the LLM decide which tool to call at every step.
# It's the "classic" agentic pattern: perceive → think → act → repeat.
#
# This version is tuned for Anna Meier (easy persona) but the
# pattern generalises to harder personas with better prompts.
# ═══════════════════════════════════════════════════════════════════════

import json
import time
import re


# ─────────────────────────────────────────────────────────────
# SYSTEM PROMPT — tells the LLM what it is and how to act
# ─────────────────────────────────────────────────────────────

AGENT_SYSTEM_PROMPT = """You are an expert Swiss tax accountant AI agent. Your job is to autonomously file a Zurich tax return (Steuererklärung 2025) for a taxpayer.

## Your workflow
1. First, call `list_documents` to see what files the taxpayer has.
2. Read each document with `read_document` to understand the taxpayer's situation.
3. Optionally call `list_guides` and `fetch_guide` to look up tax rules.
4. Call `scan_page` to see what fields are on the current form page.
5. Call `fill_field` for each field you want to fill.
6. Call `click_element` with locator "btn-next" to go to the next page.
7. Repeat steps 4-6 for each page.
8. When you reach the review page, call `submit_form`.

## Available tools
- `list_documents` — no args. Returns list of filenames.
- `read_document` — args: {"filepath": "filename.txt"}. Returns document content.
- `list_guides` — no args. Returns available tax guide topics.
- `fetch_guide` — args: {"url": "path/to/guide.html"}. Returns guide content.
- `scan_page` — no args. Returns current page name, all fields (with locators, labels, values), and buttons.
- `fill_field` — args: {"locator": "field-id", "value": "the value"}. Fills a form field.
- `click_element` — args: {"locator": "button-id"}. Clicks a button.
- `submit_form` — no args. Submits the completed tax return.
- `ask_user` — args: {"question": "your question"}. Asks the taxpayer. USE SPARINGLY — penalised if the answer was in the documents.

## Navigation IDs
- "btn-next" — the Next button on every form page
- "tab-nav-form" — switches to the form view from overview
- "btn-add-row-{section}-{table}" — adds a row to a dynamic table

## Response format
You MUST respond with valid JSON and nothing else. No markdown, no explanation outside the JSON.

```json
{
  "thought": "Brief reasoning about what to do next and why",
  "tool": "tool_name",
  "args": {"key": "value"}
}
```

If you are done (after submitting), respond:
```json
{"thought": "Tax return submitted successfully.", "tool": "done", "args": {}}
```

## Tax rules you should know
- Lohnausweis field 1 = Bruttolohn (gross salary)
- Lohnausweis field 9 = AHV/IV/EO contributions
- Lohnausweis field 10 = BVG contributions
- Lohnausweis field 13 = Spesen (already reimbursed — do NOT declare as income or deduction)
- Lohnausweis field 14.1 = Kantinenverpflegung (canteen): "Ja" → meal deduction CHF 1,600; "Nein" → CHF 3,200
- Berufsauslagen Pauschale = 3% of net salary (Bruttolohn - AHV - BVG), min CHF 2,000, max CHF 4,000
- Pillar 3a maximum = CHF 7,056 (employed with pension fund)
- Bank interest from "Zinsabschluss" entries = investment income
- Donations found in bank statement (e.g., "Spende SRK") = deductible if ≥ CHF 100 total
- Commuting: use ZVV receipt amount for Fahrkosten deduction
- Bank balance on Dec 31 (last line of CSV) = wealth to declare

## Critical rules
- Do NOT fill fields with "0", "None", or empty strings — leave irrelevant fields untouched.
- Do NOT ask the user questions if the answer is in the documents.
- When you scan a page, carefully read the field LABELS to understand what each field expects.
- Only use locators you see in scan_page results — never guess field IDs.
- After filling all fields on a page, click "btn-next" to proceed.
- On the review page, call submit_form.
"""


# ─────────────────────────────────────────────────────────────
# RESPONSE PARSER — extracts JSON from the LLM's response
# ─────────────────────────────────────────────────────────────

def parse_llm_response(text):
    """Extract a JSON action from the LLM's response text.

    Handles:
    - Clean JSON
    - JSON wrapped in ```json ... ``` code fences
    - JSON with trailing text after the closing brace
    """
    # Try to find JSON in code fences first
    fence_match = re.search(r'```(?:json)?\s*(\{.*?\})\s*```', text, re.DOTALL)
    if fence_match:
        try:
            return json.loads(fence_match.group(1))
        except json.JSONDecodeError:
            pass

    # Try to find raw JSON
    brace_match = re.search(r'\{.*\}', text, re.DOTALL)
    if brace_match:
        try:
            return json.loads(brace_match.group(0))
        except json.JSONDecodeError:
            pass

    # Fallback: ask the LLM to scan the page (safe default)
    print(f"  [WARN] Could not parse LLM response, defaulting to scan_page")
    return {"thought": "Could not parse response, scanning page.", "tool": "scan_page", "args": {}}


# ─────────────────────────────────────────────────────────────
# THE AGENTIC LOOP
# ─────────────────────────────────────────────────────────────

def think(server, client, system_prompt=None, max_steps=80):
    """Fully agentic think loop — LLM decides every action.

    Parameters
    ----------
    server : MCPServer
        The tool server with all 9 tools.
    client : OpenAI
        The LLM client.
    system_prompt : str | None
        Override system prompt. Uses AGENT_SYSTEM_PROMPT if None.
    max_steps : int
        Safety limit.
    """
    system_prompt = system_prompt or AGENT_SYSTEM_PROMPT
    conversation_history = []
    step = 0

    # State tracking — prevents the agent from repeating itself
    documents_read = []
    guides_read = []
    questions_asked = []
    fields_filled = {}        # {locator: value}
    pages_visited = []
    has_listed_documents = False
    has_listed_guides = False
    last_scan = None          # most recent scan_page result

    while step < max_steps:
        step += 1
        print(f"\n{'='*50}")
        print(f"Step {step}/{max_steps}")
        print(f"{'='*50}")

        # ─── Build the user message with current state ────
        if not has_listed_documents:
            # Very first step: force the agent to list documents
            user_message = (
                "You have just started filing this taxpayer's tax return. "
                "You don't know anything about them yet.\n\n"
                "Action required: call `list_documents` to see what files are in their folder."
            )
        else:
            # Build a state summary so the LLM knows what's been done
            state_lines = []
            state_lines.append(f"Documents read so far: {documents_read if documents_read else 'None yet'}")
            state_lines.append(f"Guides read so far: {guides_read if guides_read else 'None yet'}")
            state_lines.append(f"Questions asked: {questions_asked if questions_asked else 'None'}")
            state_lines.append(f"Pages visited: {pages_visited if pages_visited else 'None yet'}")

            if fields_filled:
                filled_summary = "\n".join(f"  {k} = {v}" for k, v in fields_filled.items())
                state_lines.append(f"Fields filled so far:\n{filled_summary}")
            else:
                state_lines.append("Fields filled so far: None yet")

            # Include last scan if available
            scan_section = ""
            if last_scan:
                scan_section = f"\n\n[CURRENT PAGE: {last_scan.get('page_name', 'unknown')}]\n"
                scan_section += json.dumps(last_scan, indent=2)[:4000]

            user_message = (
                f"[CURRENT STATE]\n"
                f"{chr(10).join(state_lines)}"
                f"{scan_section}\n\n"
                f"What is your next action? Respond with JSON only."
            )

        conversation_history.append({"role": "user", "content": user_message})

        # ─── Call the LLM ─────────────────────────────────
        time.sleep(3)  # rate limit protection

        for attempt in range(3):
            try:
                response = client.chat.completions.create(
                    model=LLM_MODEL,
                    max_tokens=1024,
                    messages=[
                        {"role": "system", "content": system_prompt}
                    ] + conversation_history[-20:]  # keep last 20 messages to manage context
                )
                break
            except Exception as e:
                if '429' in str(e) and attempt < 2:
                    print(f"  ⏳ Rate limited, waiting 30s...")
                    time.sleep(30)
                else:
                    raise

        llm_text = response.choices[0].message.content
        conversation_history.append({"role": "assistant", "content": llm_text})

        # ─── Parse the action ─────────────────────────────
        action = parse_llm_response(llm_text)

        thought = action.get("thought", "No reasoning provided.")
        tool = action.get("tool", "scan_page")
        args = action.get("args", {})

        print(f"🧠 Thought: {thought}")
        print(f"🔧 Tool:    {tool}({json.dumps(args)})")

        # ─── Handle "done" ────────────────────────────────
        if tool == "done":
            print("\n✅ Agent signaled done.")
            return {
                "filled_history": fields_filled,
                "log": [],
                "submission": None,
            }

        # ─── Handle "submit_form" ─────────────────────────
        if tool == "submit_form":
            result = server.submit_form()
            print(f"\n📤 Submitted! Success: {result.get('success', False)}")
            return {
                "filled_history": fields_filled,
                "log": [],
                "submission": result,
            }

        # ─── Execute the tool ─────────────────────────────
        try:
            result = server.execute({"tool": tool, "args": args})
        except Exception as e:
            print(f"  ❌ Error: {e}")
            result = {"error": str(e)}
            conversation_history.append({
                "role": "user",
                "content": f"Tool error: {str(e)}\nPlease try a different action."
            })
            continue

        # ─── Print result (shortened) ─────────────────────
        result_str = json.dumps(result, indent=2, default=str)
        if len(result_str) > 500:
            print(f"  📋 Result: {result_str[:500]}...")
        else:
            print(f"  📋 Result: {result_str}")

        # ─── Update state tracking ────────────────────────
        if tool == "list_documents":
            has_listed_documents = True

        elif tool == "read_document":
            fp = args.get("filepath", "")
            if fp and fp not in documents_read:
                documents_read.append(fp)

        elif tool == "list_guides":
            has_listed_guides = True

        elif tool == "fetch_guide":
            url = args.get("url", "")
            if url and url not in guides_read:
                guides_read.append(url)

        elif tool == "ask_user":
            q = args.get("question", "")
            if q:
                questions_asked.append(q)

        elif tool == "scan_page":
            last_scan = result
            page_name = result.get("page_name", "unknown") if isinstance(result, dict) else "unknown"
            if page_name and page_name not in pages_visited:
                pages_visited.append(page_name)

        elif tool == "fill_field":
            if isinstance(result, dict) and result.get("success"):
                fields_filled[args.get("locator", "")] = args.get("value", "")

        elif tool == "click_element":
            if isinstance(result, dict) and result.get("new_page"):
                last_scan = None  # force a fresh scan on the new page

        # ─── Feed result back to the LLM ──────────────────
        # Truncate large results (e.g., full document text) to
        # keep the context window manageable
        result_for_llm = json.dumps(result, indent=2, default=str)
        if len(result_for_llm) > 6000:
            result_for_llm = result_for_llm[:6000] + "\n... (truncated)"

        conversation_history.append({
            "role": "user",
            "content": f"Tool `{tool}` result:\n{result_for_llm}"
        })

    print(f"\n⚠️ Max steps ({max_steps}) reached without submitting.")
    return {
        "filled_history": fields_filled,
        "log": [],
        "submission": None,
    }
````
